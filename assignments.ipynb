2. What is an interpreter in Python ?

Ans. An interpreter in Python is a program that reads and executes Python code line by line. Instead of compiling the entire program into machine code before running it (like a compiler does), the interpreter translates each Python statement into machine code and executes it immediately.

3. What are pre-defined keywords in Python ?

Ans. Pre-defined keywords in Python are reserved words that have special meaning and syntax in the language. These keywords cannot be used as identifiers (such as variable names, function names, or class names) because they are part of the Python language syntax.

4. Can keywords be used as variable names?

Ans. No, keywords cannot be used as variable names in Python. because keywords have special meanings and are reserved by the language, using them as variable names will result in a syntax error.

5. What is mutability in Python?

Ans. Mutability in Python refers to whether an object’s value can be changed after it is created.

  * Mutable objects can be modified after creation. You can change their content without creating a new object.

 * Immutable objects cannot be changed once created. Any modification results in a new object being created.

 9. What is type casting in Python?

 Ans. Type casting in Python is the process of converting a value from one data type to another. This is useful when you need to perform operations that require specific data types or when you want to ensure data is in the correct format.

 10. What is the difference between implicit and explicit type casting?

 Ans. 1. Implicit Type Casting (Type Coercion)
 * Performed automatically by Python without programmer intervention.
 * Happens when Python automatically converts one data type to another to avoid data loss or errors during operations.
 * Usually occurs in expressions involving mixed data types.

   2.  Explicit Type Casting (Type Conversion)
      * Performed manually by the programmer using built-in functions.
      * Requires the programmer to explicitly specify the target data type.
      * Used when you want to convert data types intentionally.  

 11. What is the purpose of conditional statements in Python?

 Ans. In Python, conditional statements are used to make decisions in  program by executing certain blocks of code only when specific conditions are true.

 12. How does the elif statement work?

 Ans. the elif statement is short for "else if" and is used to check multiple conditions sequentially after an initial if statement. It allows you to test several expressions and execute a block of code as soon as one of the conditions evaluates to True.

 13. What is the difference between for and while loops

 Ans.  'for 'loop
   * Used to iterate over a sequence (like a list, tuple, string, or range).
   * The number of iterations is usually known or finite.
   * It automatically moves through each item in the sequence.

   'while' loop
   * Repeats as long as a condition is True.
   * The number of iterations is not necessarily known beforehand.
   * You control the loop continuation explicitly with a condition.

```python
# 1. Write a Python program to print "Hello, World"

print("Hello, World!")

# 2.Write a Python program that displays your name and age.

name = "Ankit Patel"
age = 18

print("Name:", name)
print("Age:", age)

# 3. Write code to print all the pre-defined keywords in Python using the keyword library.

import keyword

print(keyword.kwlist)

# 4. Write a program that checks if a given word is a Python keyword.

import keyword

word = input("Enter a word: ")

if keyword.iskeyword(word):
    print(f'"{word}" is a Python keyword.')
else:
    print(f'"{word}" is not a Python keyword.')

# 5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
for each.

# Creating a list
my_list = [1, 2, 3]
print("Original list:", my_list)

my_list[1] = 20
print("Modified list:", my_list)


my_tuple = (1, 2, 3)
print("Original tuple:", my_tuple)


try:
    my_tuple[1] = 20
except TypeError as e:
    print("Error when trying to modify tuple:", e)

# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.

def modify_arguments(immutable_arg, mutable_arg):
    print("Inside function before modification:")
    print("Immutable argument:", immutable_arg)
    print("Mutable argument:", mutable_arg)

    # Attempt to modify the immutable argument (e.g., an integer)
    immutable_arg += 10

    # Modify the mutable argument (e.g., a list)
    mutable_arg.append(100)

    print("Inside function after modification:")
    print("Immutable argument:", immutable_arg)
    print("Mutable argument:", mutable_arg)



immutable_value = 5
mutable_value = [1, 2, 3]

print("Before function call:")
print("Immutable value:", immutable_value)
print("Mutable value:", mutable_value)

modify_arguments(immutable_value, mutable_value)

print("After function call:")
print("Immutable value:", immutable_value)  # unchanged
print("Mutable value:", mutable_value)      # changed

# 7. Write a program that performs basic arithmetic operations on two user-input numbers.

# Get two numbers from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Perform arithmetic operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2


if num2 != 0:
    division = num1 / num2
else:
    division = "Undefined (division by zero)"

# Display the results
print(f"{num1} + {num2} = {addition}")
print(f"{num1} - {num2} = {subtraction}")
print(f"{num1} * {num2} = {multiplication}")
print(f"{num1} / {num2} = {division}")

# 8. Write a program to demonstrate the use of logical operators.

# Input values
a = True
b = False

print("a =", a)
print("b =", b)

# Using 'and' operator
print("\nLogical AND (a and b):", a and b)

# Using 'or' operator
print("Logical OR (a or b):", a or b)

# Using 'not' operator
print("Logical NOT (not a):", not a)
print("Logical NOT (not b):", not b)


x = 10
y = 5

print("\nExample with numbers:")
print("x > 0 and y > 0:", x > 0 and y > 0)
print("x > 0 or y < 0:", x > 0 or y < 0)
print("not(x > y):", not(x > y))
 Sample Output:
yaml
Copy code
a = True
b = False

Logical AND (a and b): False
Logical OR (a or b): True
Logical NOT (not a): False
Logical NOT (not b): True

Example with numbers:
x > 0 and y > 0: True
x > 0 or y < 0: True
not(x > y): False

# 9. Write a Python program to convert user input from string to integer, float, and boolean types.

# Taking input from the user
user_input = input("Enter a value: ")

# Convert to integer
try:
    int_value = int(user_input)
    print("Integer value:", int_value)
except ValueError:
    print("Cannot convert to integer.")


try:
    float_value = float(user_input)
    print("Float value:", float_value)
except ValueError:
    print("Cannot convert to float.")

bool_value = bool(user_input)
print("Boolean value:", bool_value)

# 10. Write code to demonstrate type casting with list elements.

# A list of string numbers
str_list = ["10", "20", "30", "40", "50"]

print("Original list (strings):", str_list)

# Convert string elements to integers
int_list = [int(x) for x in str_list]
print("List after converting to integers:", int_list)

# Convert integers to float
float_list = [float(x) for x in int_list]
print("List after converting to floats:", float_list)

# Convert integers to boolean
bool_list = [bool(x) for x in int_list]
print("List after converting to booleans:", bool_list)

# Convert integers back to strings
back_to_str_list = [str(x) for x in int_list]
print("List after converting back to strings:", back_to_str_list)

# 11. Write a program that checks if a number is positive, negative, or zero.

# Take input from the user
num = float(input("Enter a number: "))

# Check conditions
if num > 0:
    print("The number is Positive.")
elif num < 0:
    print("The number is Negative.")
else:
    print("The number is Zero.")

# 12. Write a for loop to print numbers from 1 to 10.

# Program to print numbers from 1 to 10 using a for loop

for i in range(1, 11):
    print(i)

# 13. Write a Python program to find the sum of all even numbers between 1 and 50.

total = 0

for i in range(1, 51):
    if i % 2 == 0:   # check if number is even
        total += i

print("The sum of even numbers between 1 and 50 is:", total)

# 14. Write a program to reverse a string using a while loop.

# Take input from user
text = input("Enter a string: ")

# Initialize variables
reversed_text = ""
i = len(text) - 1   # start from the last index

# Loop to reverse the string
while i >= 0:
    reversed_text += text[i]
    i -= 1

print("Reversed string:", reversed_text)

# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

# Take input from user
num = int(input("Enter a number: "))

factorial = 1
i = 1

# Factorial calculation using while loop
while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)







```
